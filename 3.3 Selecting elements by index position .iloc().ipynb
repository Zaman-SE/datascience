```python
import pandas as pd
```

```python
#read csv file
df = pd.read_csv('players_20.csv', encoding='latin1')
#set index
df.set_index('short_name', inplace=True)
#select columns
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
#show dataframe
df
```

# 1 Selecting with a single value
iloc[row_position, column_position]

```python
#get the height of Cristiano
df.iloc[0, 3]
```

```python
#get the weight of Cristiano
df.iloc[0, 4]
```

```python
#get all the rows inside the 'height_cm' column
df.iloc[:, 3]
```

```python
#get all the columns that corresponds to 'Cristiano'
df.iloc[0, :]
```

# 2 Selecting with a list of values

```python
#get all data about messi and Cristiano
df.iloc[[0, 1]]
```

```python
#get the height of Cristiano and Messi
df.iloc[[0, 1], 3]
```

```python
#get the height and weight of Cristiano
df.iloc[[0], [3, 4]]
```

```python
#get the height and weight of Cristiano and Messi
df.iloc[[0, 1], [3, 4]]
```

# 3 Selecting a range of data with slice
start:stop:step

```python
#slice column labels
df.iloc[[0, 1], 2 : 5]
```

```python
#slice index Labels

#get top1 and top10 player name
df.index[:10]
```

```python
df.iloc[0:10, 2:5]
```

# Selecting with Conditions

```python
#one condition: selecting player with height above 180 cm
df.iloc[list(df['height_cm']>180)]
```

```python
#multiple condition: selecting player with height above 180 cm and from france
df.iloc[list((df['height_cm']>180) & (df['nationality'] == 'France')), :]
```
