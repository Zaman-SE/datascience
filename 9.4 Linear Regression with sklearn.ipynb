```python
df_boston = pd.read_csv('Boston House Prices.csv')
df_boston
```

# 1 Linear Regression with Scikit-learn


```python
from sklearn import linear_model
```

## 1.1 Define the dependent and independent variable

```python
#dependent variable
y=df_boston['Value']
#independent variable
x=df_boston[['Rooms', 'Distance']]
```

## 1.2 Fitting the model

```python
lm = linear_model.LinearRegression()
lm.fit(x, y)
```

## 1.3 Predicting values

```python
lm.predict(x)[:5] #getting first 5 values
```

## 1.4 Regression table

```python
#r2 score
lm.score(x, y)
#coefficient
lm.coef_
#intercept
#lm.intercept_
```

# 1.5 Linear Regression Equation

```python
#linear equation: y=ax+b
y_pred = 8.80*x['Rooms'] + 0.49*x['Distance'] -34.63
```

### 1.1.6 Plotting the Regression Line


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# grid
rooms = np.linspace(df_boston['Rooms'].min(),
                    df_boston['Rooms'].max(), 20)
distance = np.linspace(df_boston['Distance'].min(),
                       df_boston['Distance'].max(), 20)

R, D = np.meshgrid(rooms, distance)

# predict
Y = lm.intercept_ + lm.coef_[0]*R + lm.coef_[1]*D

# plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_boston['Rooms'], df_boston['Distance'], y)
ax.plot_surface(R, D, Y, alpha=0.5)

ax.set_xlabel("Rooms")
ax.set_ylabel("Distance")
ax.set_zlabel("Value")

plt.show()
```
