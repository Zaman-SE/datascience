```python
import pandas as pd
```

```python
#reading the csv file
df = pd.read_csv('population_total.csv')
```

```python
#first look at the dataset
df
```

```python
#dropping null values
df.dropna(inplace=True)
```

```python
#making a pivot table
df = df.pivot(index='year', columns='country', values='population')
```

```python
#selecting some countries
#df = df[['Bangladesh', 'United States', 'China', 'Indonesia']]
```

```python
#showing pivot table
df
```

# 1 lineplot

```python
df.plot(kind='line', xlabel='Year', ylabel='Population', title='Population (1955 - 2020)', figsize=(8, 4))
```

# 3 Barplot
## 3.1 Single Barplot

```python
#selecting only one year(2020)
df_one_year = df[df.index.isin([2020])]
```

```python
#showing new dataframe
df_one_year
```

```python
#transpose dataframe from rows to columns or vice versa
df_transpose = df_one_year.T
```

```python
# making barplot
df_transpose.plot(kind='bar')
```

## 3.2 Barplot grouped by "n" variables

```python
#selecting a few years
df_multiple_year = df[df.index.isin([1960, 1970, 1980, 1990, 2020])]
```

```python
#making grouped barplot
df_multiple_year.plot(kind='bar')
```

# 4 piechart

```python
#Changing column name
df_one_year.rename(index={2020:'2020'}, inplace=True)
```

```python
#making the plot
df_one_year = df_one_year.T
df_one_year
```

```python
df_one_year.plot(kind='pie', y='2020')
```
