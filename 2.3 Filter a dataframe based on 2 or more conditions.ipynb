```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
```

```python
df_laptops.head(3)
```

# 1. Filter dataframe based on a condition

```python
#find apple laptops
df_laptops['Company'] == 'Apple'
```

```python
#find laptops higher than 2000 euros
df_laptops['Price_euros'] > 2000
```

# 2. Filter Dataframe based on multiple conditions

```python
#find Apple laptops that costs more than 2000 euros
(df_laptops['Company'] == 'Apple') & (df_laptops['Price_euros'] > 2000)
```

## 2.1 &

```python
#filter dataframe based on multiple conditions
df_laptops[(df_laptops['Company'] == 'Apple') & (df_laptops['Price_euros'] > 2000)]
```

## 2.2 |

```python
#find apple or dell laptops
(df_laptops['Company'] == 'Apple' ) | (df_laptops['Company'] == 'Dell')
```

```python
#filter dataframe based on multiple conditions
df_laptops[(df_laptops['Company'] == 'Apple' ) | (df_laptops['Company'] == 'Dell')]
```

```python
#find laptops from Apple or Dell that costs more than 2000 euros
((df_laptops['Company'] == 'Apple' ) | (df_laptops['Company'] == 'Dell')) & (df_laptops['Price_euros'] > 2000)
```

```python
#filter dataframe based on multiple conditions
df_laptops[(df_laptops['Company'] == 'Apple' ) | (df_laptops['Company'] == 'Dell')]
```
