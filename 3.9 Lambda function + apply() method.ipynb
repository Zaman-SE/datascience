```python
import pandas as pd
```

```python
#reads csv file
df = pd.read_csv('players_20.csv', encoding='latin1')
#set index
df.set_index('short_name', inplace=True)
#select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
#show dataframe
df
```

# 1 lambda function

```python
#basic function
def sum_values(a, b):
    x = a+b
    return x
```

```python
sum_values(2, 4)
```

```python
#lambda function
sum_values = lambda a,b:a+b
```

```python
sum_values(2, 3)
```

# 2 Apply + Lambda function

```python
#use lambda function to convert 'height_cm' to meters
df['height_cm'].apply(lambda x:x/100)
#alternative
df['height_cm']/100
```

```python
#use lambda function to convert 'long name' series to upper case
df['long_name'].apply(lambda x:x.upper())
#alternative with str attribute
df['long_name'].str.upper()
```

```python
df['dob'] = df['dob'].astype('datetime64[ns]')
#use lambda function to get the year of 'dob' series
df['dob'].apply(lambda x:x.year)
#alternative with dt attribute
df['dob'].dt.year
```

```python
#apply lambda function to dataframe in order to calculate bmi
lambda x:x['weight_kg']/((x['height_cm'])**2)
df.apply(lambda x:x['weight_kg']/((x['height_cm']/100)**2), axis=1)
```
