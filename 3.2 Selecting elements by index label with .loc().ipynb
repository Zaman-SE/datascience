```python
import pandas as pd
```

```python
#read csv file
df = pd.read_csv('players_20.csv', encoding='latin1')
#set index
df.set_index('short_name', inplace=True)
#select columns
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
#show dataframe
df
```

# 1. Selecting with a single value
loc[row_label, column_label]

```python
#get all data about Cristiano
df.loc['Cristiano Ronaldo']
```

```python
#get the height of Cristiano
int(df.loc['Cristiano Ronaldo', 'height_cm'])
```

```python
#get the weight of Cristiano
int(df.loc['Cristiano Ronaldo', 'weight_kg'])
```

```python
#get all the rows inside the 'height_cm' column
df.loc[:, 'height_cm']
```

```python
#get all columns that corresponds to the index 'Cristiano Ronaldo'
df.loc['Cristiano Ronaldo', :]
```

# 2 Selecting with a list of values

```python
#get all data about Cristiano and Messi
df.loc[['Cristiano Ronaldo', 'L. Messi']]
```

```python
#get the height of Cristiano and Messi
df.loc[['Cristiano Ronaldo', 'L. Messi'], 'height_cm']
```

```python
#get the height and weight of Cristiano
df.loc['Cristiano Ronaldo', ['height_cm', 'weight_kg']]
```

```python
#get the height and weight of Cristiano and Messi
df.loc[['Cristiano Ronaldo', 'L. Messi'], ['height_cm', 'weight_kg']]
```

# 3 Selecting a range of data with slice
start:stop:step

```python
#slice column labels
players = ['Cristiano Ronaldo', 'L. Messi']
df.loc[players, 'age':'club']
```

```python
#slice index Labels
columns = ['age', 'dob', 'height_cm', 'weight_kg']
#get top1 and top10 player name
df.index[:10]
df.loc['Cristiano Ronaldo':'M. Salah', columns]
```

# 4 Selecting with Conditions

```python
#one condition: selecting player with height above 180 cm
columns = ['age', 'dob', 'height_cm', 'weight_kg']
df.loc[df['height_cm']>180, columns]
```

```python
#multiple condition: selecting player with height above 180 cm and from france
df.loc[(df['height_cm']>180) & (df['nationality']=='Portugal'), columns]
```
