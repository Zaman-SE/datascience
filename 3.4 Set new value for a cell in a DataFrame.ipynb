```python
import pandas as pd
```

```python
#read csv file
df = pd.read_csv('players_20.csv', encoding='latin1')
#set index
df.set_index('short_name', inplace=True)
#select columns
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
#show dataframe
df
```

# 1 Set Value to one cell

```python
#update Cristiano height
df.loc['Cristiano Ronaldo', 'height_cm'] = 185
```

```python
#show dataframe
df
```

# 2 Set value to entire column

```python
#set height of all players to 190
df.loc[:, 'height_cm'] = 185
```

```python
#show dataframe
df
```

# 3 set value to entire row

```python
#get all columns correspond to player ranked last in fifa
df.iloc[-1, :]
```

```python
#set null values to all columns that corresponds to player ranked last
import numpy as np
df.iloc[-1, :] = np.nan
```

```python
#show dataframe
df
```

# 4 Set value to multiple cells

```python
#set value for all items matching the list of labels
df.loc[['Cristiano Ronaldo', 'L. Messi'], 'height_cm'] = 170
```

```python
#show dataframe
df
```

# 5 Set value for rows matching a condition

```python
#set value for rows matching a condition
df.loc[df['height_cm']>180, :] = 0
```

```python
#show dataframe
df
```
