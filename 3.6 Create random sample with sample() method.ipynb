```python
import pandas as pd
```

```python
#reads csv file
df = pd.read_csv('players_20.csv', encoding='latin1')
#set index
df.set_index('short_name', inplace=True)
#select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
#show dataframe
df
```

# 1 sample()

```python
#extract 10 random elements from the 'nationality' column
df['nationality'].sample(10, random_state=4)
```

```python
#extract a random 20% sample of the dataframe
df.sample(frac=.2, random_state=20)
```

```python
#upsample: increase the sampling rate -> frac>1 (Note: replacement parameter)
df.sample(frac=2, replace=True)
```
