```python
import pandas as pd
```

```python
df_movies = pd.read_csv('IMDb movies.csv', low_memory=False)
df_ratings = pd.read_csv('IMDb ratings.csv')
```

```python
df_ratings
```

```python
#select columns
df_movies = df_movies[['imdb_title_id', 'title', 'year', 'genre', 'country']]
df_ratings = df_ratings[['imdb_title_id', 'total_votes', 'mean_vote']]
```

# 1 merge()

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'id': ['C', 'D', 'E', 'F'],
                    'job': ['Doctor', 'Engineer', 'Accountant', 'Developer']})
```

```python
df1
```

```python
df2
```

## 1.1 Inner join

```python
df_inner_join = df1.merge(df2, on='id', how='inner')
df_inner_join
```

```python
#shape
print(df1.shape)
print(df2.shape)
print(df_inner_join.shape)
```

# 1.2 Outer join(Full join)

```python
df_outer_join = df1.merge(df2, on='id', how='outer')
df_outer_join
```

## 1.2.1 Exercise

```python
#merge df_movies and df_ratings(outer join)
df_outer_join = df_movies.merge(df_ratings, on='imdb_title_id', how='outer')
```

```python
df_outer_join
```

# 1.3 Exclusive outer join(Exclusive full join)

```python
df_outer_join = df1.merge(df2, on='id', how='outer')
df_outer_join
```

```python
#indicator=True
df_outer_join = df1.merge(df2, on='id', how='outer', indicator=True)
df_outer_join
```

```python
#query
df_outer_join = df1.merge(df2, on='id', how='outer', indicator=True).query("_merge=='left_only' or _merge=='right_only'")
df_outer_join
```

```python
#shape
df_outer_join.shape
```

## 1.5.1 Exercise

```python
#make a copy of the df_movies dataframe
df_movies_2 = df_movies.copy()
df_movies_2
```

```python
#set the first 1000 values of 'imdb_title_id' column as 'tt1234567890'
for index in df_movies_2.index:
    if index<1000:
        df_movies_2.loc[index, 'imdb_title_id'] =   'tt1234567890'
```

```python
df_movies_2
```

```python
#merge df_movies_2 and df_ratings(exclusive left join)
df_exclusive_left = df_movies_2.merge(df_ratings, on='imdb_title_id', how='outer', indicator=True).query("_merge=='left_only'")
df_exclusive_left
```

```python
#shape
df_exclusive_left.shape
```
