```python
import pandas as pd
```

```python
#reads csv file
df = pd.read_csv('players_20.csv', encoding='latin1')
#set index
df.set_index('short_name', inplace=True)
#select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
#show dataframe
df
```

# 1 copy()
## 1.1 deep=True

```python
#deep=True by default(Modifications to the data orindices of the copy)
df_copy = df.copy()
```

```python
#update value in original DataFrame
df.loc['Cristiano Ronaldo', 'height_cm'] = 195
```

```python
#copy vs original DataFrame
df
df_copy
```

## 1.2 deep=False

```python
#deep=False by default(Any changes to the data of the original will be reflected in the shallow copy and vice versa)
df_copy = df.copy(deep=False)
```

```python
#update value in original DataFrame
df.loc['Cristiano Ronaldo', 'height_cm'] = 195
```

```python
#shallow copy vs original DataFrame
df
#df_copy
```

# 1.3 Simple assignment

```python
#make a copy
df_copy = df
```

```python
#update value in original DataFrame
df.loc['Neymar Jr', 'height_cm'] = 170
```

```python
#copy vs original dataframe
df_copy
```
