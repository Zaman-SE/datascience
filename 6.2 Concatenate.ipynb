```python
import pandas as pd
```

```python
df_movies = pd.read_csv('IMDb movies.csv', low_memory=False)
df_ratings = pd.read_csv('IMDb ratings.csv')
```

```python
#select columns
df_movies = df_movies[['imdb_title_id', 'title', 'year', 'genre', 'country']]
df_ratings = df_ratings[['imdb_title_id', 'total_votes', 'mean_vote']]
```

```python
df_movies
```

# 1 Concat
## 1.1 Concatenate vertically

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'id': ['E', 'F', 'G', 'H'],
                    'age': [40, 21, 19, 24]})
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True)
```

## 1.1.1 Exercise

```python
#extract a 50% sample of the original dataframe(df_movies)
df_sample = df_movies.sample(frac=0.5)
df_sample
```

```python
#shape of dataframe that we contcatenate
df_sample.shape
```

```python
#concatenate df movies and df_sample
df_concat_vertical = pd.concat([df_movies, df_sample])
```

## 2.1 Concatenate horizontally

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'job': ['Doctor', 'Engineer', 'Accountant', 'Developer']})
```

```python
df_concat_horizontal = pd.concat([df1, df2], axis=1)
df_concat_horizontal
```

## 1.2.1 Exercise

```python
df_movies.set_index('imdb_title_id', inplace=True)
df_ratings.set_index('imdb_title_id', inplace=True)
```
