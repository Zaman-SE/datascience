```python
import pandas as pd
```

```python
#reads csv file
df = pd.read_csv('players_20.csv', encoding='latin1')
#set index
df.set_index('short_name', inplace=True)
#select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
#show dataframe
df
```

# 1 Drop row

```python
#drop one row
#axis parameter
df.drop('L. Messi', axis=0)
#index parameter
df.drop(index=['L. Messi'])
```

```python
#drop two or more rows and update date(inplace=True)
df.drop(df.index[1:4], inplace=True)
```

```python
df
```

# 2 Drop columns

```python
#drop one column
#axis parameter
df.drop('dob', axis=1)
#index parameter
df.drop(columns='dob')
```

```python
#drop two or more rows and update date(inplace=True)
df.drop(['long_name', 'age'], axis = 1, inplace=True)
```
