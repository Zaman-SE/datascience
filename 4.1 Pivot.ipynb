# 2 Pivot Table

```python
#read and show the dataset
df = pd.read_excel('supermarket_sales.xlsx')
df
#df.plot(kind='line')
#strip any hidden 
```

```python
#clean column names 
df.columns = df.columns.str.strip()
#Total sales by Gender
df.pivot_table(values='Total', index='Gender', aggfunc='sum', margins=True)
```

```python
#Total sales by City and Gender
df.pivot_table(values='Total', index='City', columns='Gender', aggfunc='sum', margins=True)
```

```python
#Average rating by Product line and Customer Type
df.pivot_table(values='Rating', index='Product line', columns='Customer type', aggfunc='mean')
```

```python
#select some columns
df.pivot_table(index='Gender', values=['Quantity', 'Total'], aggfunc='sum')
```
